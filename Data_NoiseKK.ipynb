```python
def interpolate_noiseKK(l_min, l_max):
    
    ellK = np.linspace(l_min, l_max, (l_max-l_min+1))
    nlK = np.interp(ellK, ell, nlkk)
    
    noise_K = np.zeros(l_max+1, 'float64')
    for l in range(l_min, l_max+1):
        noise_K[l] = nlK[l-l_min]
    return noise_K
```

```python
noiseKK = interpolate_noiseKK(0, 1500)
plt.figure
plt.yscale('log')
plt.plot(ell,nlkk)
plt.plot(noiseKK, 'r')
```
